```python
def download_file(url: str, file_path):
  """
  download file and save on file path
  """
  file_content = requests.get(url).text
  file = open(file_path, 'w')
  file.write(file_content)
  file.close()
```

```python
download_file(answer_url, answer_file_path)
download_file(question_url, questions_file_path)
```

#### is Ethnicity list?
بررسی قومیت

```python
ethnicity = list(answer_df['Ethnicity'].unique())
ethnicity = [x.split(";") for x in ethnicity if isinstance(x, str)]
all_ethnicity = list()
for each in ethnicity:
  all_ethnicity += each
set(all_ethnicity)
```

یک فرد اگر بیشتر از دو تا قومیت انتخاب کرده باشد چند نژادی محسوب می‌شود اگر دو قومیت انتخاب کرده باشه دونژادی محسوب می‌شود.

```python
multi_list = [x for x in list(answer_df['Ethnicity'].unique()) if isinstance(x, str) and len(x.split(";")) > 2]
bi_list = [x for x in list(answer_df['Ethnicity'].unique()) if isinstance(x, str) and len(x.split(";")) == 2]
for each in multi_list:
  answer_df.loc[(answer_df['Ethnicity']==each), 'Ethnicity'] = "Multiracial"
for each in bi_list:
  answer_df.loc[(answer_df['Ethnicity']==each), 'Ethnicity'] = "Biracial"
```

###convert

```python
for col in list_value_col:
  #convert to list
  answer_df[col]=answer_df[col].apply(lambda x: list(x.split(';')) if not isinstance(x, float) else list())

  #create new cols
  new_cols = pd.get_dummies(answer_df[col].explode()).groupby(level=0).sum()
  new_cols.columns = [f"{col}_{str(x)}" for x in new_cols.columns]

  #join to main dataset
  answer_df=answer_df.join(new_cols)

  #delete col
  # answer_df.drop(col, inplace=True, axis=1)
```

##CompFreq and CompTotal
اطلاعات مربوط به‌کسانی است که شغل دارند. در آمد بر اساس سه ویژگی پر خواهند شد. اگر شاغل نبود می‌توان یکی از فرکوئنسی ها به همراه درآمد صفر جایگزین  شود اما به دلیل اینکه میانگین‌ها به هم می‌ریزد این کار انجام نمی‌شود.

```python
for em in employs:
  for ferq in ferqs:
    for sym in symbols:
      condition = ((answer_df['Employment']==em)&
                   (answer_df['CompFreq']==ferq)&
                   (answer_df['CurrencySymbol']==sym))
      try:
        answer_df.loc[(condition&answer_df['CompTotal'].isnull()), "CompTotal"] = answer_df[condition]['CompTotal'].mode().iloc[0]
      except:
        answer_df.loc[(condition&answer_df['CompTotal'].isnull()), "CompTotal"] = answer_df[condition]['CompTotal'].mode()
```

Age1stCode = Age - Ceil(YearsCode)
2.   اگر فقط قانون ۲ نقض شود، قانون ۴ نیز نقض می‌شود: مجموع‌ سال‌های کد زدن با استفاده از فرمول زیر تغییر می‌کند<br />
YearsCode = Age - Age1stCode
3.   اگر قانون ۴ نقض شود (در اینجا تنها می‌تواند با قانون ۳ یا نتها نقض شود) با استفاده از فرمول زیر اعداد تغییر خواهند کرد<br />
Age = Age1stCode + ceil(YearsCode) if 8 <= new_age =< 100 <br />
YearsCode = Age - Age1stCode elif not rule 3<br />
Age1stCode = Age - Ceil(YearsCode) else
4.   اگر قانون ۳ نقض شود (در اینجا می‌تواند با قانون ۴ یا با قانون ۱ یا بدون آن‌ها نقض شود): سال‌های فعالیت حرفه‌ای با استفاده از فرمول زیر دوباره محاسبه می‌شود<br />
YearsCodePro = YearsCode
5.   اگر فقط قانون‌های ۱ و ۲ نقض شود که باعث نقض ۴ نیز می‌شود محاسبات به‌صورت زیر خواهد بود<br />
Age = Age1stCode + ceil(YearsCode) if 8 <= new_age =< 100<br />
YearsCode, Age1stCode = None else
6.   اگر فقط قانون‌ها ۲ و ۳ نقض شود ابتدا با استفاده از حالت ۲ و سپس اگر قانون ۳ همچنان نقض بود حالت ۴ اجرا می‌شود<br />
