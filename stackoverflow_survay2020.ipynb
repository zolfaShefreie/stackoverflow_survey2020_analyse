```python
import requests
```

```python
answer_url = 'https://raw.githubusercontent.com/zolfaShefreie/stackoverflow_survey2020_analyse/main/Answers.csv'
question_url = "https://raw.githubusercontent.com/zolfaShefreie/stackoverflow_survey2020_analyse/main/Questions.csv"

answer_file_path = "Answers.csv"
questions_file_path = "Questions.csv"
```

```python
def download_file(url: str, file_path):
  """
  download file and save on file path
  """
  file_content = requests.get(url).text
  file = open(file_path, 'w')
  file.write(file_content)
  file.close()
```

```python
download_file(answer_url, answer_file_path)
download_file(question_url, questions_file_path)
```
